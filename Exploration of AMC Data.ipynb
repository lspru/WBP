```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_excel('All Runs.xlsx', sheet_name = 'First 27')
```

```python
df
```

```python
X = df[['SV1', 'SV2', 'SV3', 'SV4', 'SV5', 'SV6', 'SV7', 'SV8', 'SV9', 'SV10']]
y = df[['Value']]
```

```python
regr = linear_model.LinearRegression()
regr.fit(X.values, y.values)
```

```python
predicted_effic = regr.predict([[1.0, 1.5, 2.0, 1.0, 1.5, 2.0, 1.0, 1.5, 2.0, 1.0]])
print(predicted_effic)
```

```python
print(regr.coef_)
```
